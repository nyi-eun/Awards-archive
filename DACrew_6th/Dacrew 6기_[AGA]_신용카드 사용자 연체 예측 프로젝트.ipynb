ex) 성별(남,여), 혈액형(A,B,AB,O), 국적  등

>
>

## 1. gender_성별

성별의 분포를 확인하기 위해 groupby를 활용해 count를 진행해주었습니다.

```python
train.groupby('gender').count()['index']
```

시각화의 경우, pie plot을 사용하려면 plt.pie()를 사용하면 되고, 안에 원하는 색상 코드나 점선 등의 옵션을 추가적으로 사용해 plot을 꾸미는 것도 가능합니다.   
저는 핑크색을 좋아하기 때문에 pie chart의 색상은 핑크 계열로 설정하겠습니다.

## Data Cleansing

Data cleansing은 말 그대로 데이터 분석 전, 데이터를 분석에 적합한 형태로 처리하는 과정을 말합니다.

## 1. 결측치 처리

결측값이란 데이터셋 내에서 값이 누락되어 비어 있는 상태를 의미합니다.  
결측값이 있는 상태로 모델을 만들 경우, 변수 간의 관계가 왜곡될 수 있습니다.  
따라서, 결측값이 발생하는 유형, 혹은 해당 Feature의 특성에 따라 결측값을 올바르게 처리해야 합니다.

Data Cleansing의 종류로는 1. 결측치 처리 2. 이상치 처리 3. Scaling 4. Encoding로 나눌 수 있는데,  
본 stage 2에서는 결측치 및 이상치 처리만 진행하겠습니다. (Scaling 및 Encoding이 궁금하면 stage 3에서 확인해보시면 되겠습니다.)

데이터프레임에 대해 isnull().sum()을 붙여주면 데이터의 결측치 개수를 한 눈에 파악할 수 있습니다.

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

추가적으로 결측치의 퍼센트를 확인해보고 싶다면 아래의 코드를 활용해 볼 수도 있습니다.

```python
# train data 결측치 비율

for col in train.columns:
    msg = f'{col:>10} \t {100 * train[col].isnull().mean():.2f}%'
    print(msg)
```

```python
for col in test.columns:
    msg = f'{col:>10} \t {100 * test[col].isnull().mean():.2f}%'
    print(msg)
```

해당 코드는 반복문으로 데이터들의 열을 하나씩 반환하며 해당 열에서의 null값 갯수의 mean()계산으로 % 표현을 할 수 있는 코드입니다.

확인해본 결과, occyp_type에만 결측치가 존재했고, 약 30% ~ 32%라는 결측치 비율을 통해 결측치 처리를 다음과 같이 진행해주었습니다.

#### [occyp_type]

결측치 확인 및 구체적인 내용을 보기 위해 value_counts()라는 것을 붙여주면 아래와 같이 각 컬럼에 대한 값의 개수를 내림차순으로 확인할 수 있습니다.  
추가적으로 dropna=False로 지정을 해주면, NaN값까지도 자세히 확인이 가능합니다.

## 4. Feature 추가

머신러닝에서의 피처 추가는 새로운 데이터 속성을 생성하여 모델의 성능을 향상시키는 과정을 의미합니다.  
이를 통해 다음과 같은 이점을 얻을 수 있습니다.  
  
첫째, 도메인 지식을 활용하여 새로운 피처를 생성하면 모델이 데이터의 특성을 더 잘 파악할 수 있다.  
  
둘째, 유용한 정보를 담은 새로운 피처는 모델의 예측 능력을 향상시켜 정확도를 높일 수 있다.  
  
셋째, 데이터의 다양한 측면을 포착하는 피처를 추가함으로써 모델의 일반화 능력을 향상시키고 과적합을 줄일 수 있다.

피쳐 추가는 train 데이터에 적용한 함수 그대로 test 셋에 적용하여 진행합니다.

```python
import numpy as np
```

피쳐 생성을 위해 데이터 구조, 수치 계산을 위해 효율적으로 구현된 기능을 제공하는 numpy 라이브러리를 불러줍니다.

##### [before_EMPLOYED]
태어나서 고용되기 이전 시점까지의 기간을 의미하는 피쳐를 생성
- before_EMPLOYED_m : month
- before_EMPLOYED_w : week

```python
train['before_EMPLOYED'] = train['DAYS_BIRTH'] - train['DAYS_EMPLOYED']
train['income_total_befofeEMP_ratio'] = train['income_total'] / train['before_EMPLOYED']
train['before_EMPLOYED_m'] = np.floor(train['before_EMPLOYED'] / 30) - ((np.floor(train['before_EMPLOYED'] / 30) / 12).astype(int) * 12)
train['before_EMPLOYED_w'] = np.floor(train['before_EMPLOYED'] / 7) - ((np.floor(train['before_EMPLOYED'] / 7) / 4).astype(int) * 4)

test['before_EMPLOYED'] = test['DAYS_BIRTH'] - test['DAYS_EMPLOYED']
test['income_total_befofeEMP_ratio'] = test['income_total'] / test['before_EMPLOYED']
test['before_EMPLOYED_m'] = np.floor(test['before_EMPLOYED'] / 30) - ((np.floor(test['before_EMPLOYED'] / 30) / 12).astype(int) * 12)
test['before_EMPLOYED_w'] = np.floor(test['before_EMPLOYED'] / 7) - ((np.floor(test['before_EMPLOYED'] / 7) / 4).astype(int) * 4)
```

##### [Age], [DAYS_BIRTH_m], [DAYS_BIRTH_w]
Age(나이), 태어난 월, 태어난 주(출생연도의 n주차) 피쳐 생성
- DAYS_BIRTH_m : month
- DAYS_BIRTH_w : week

```python
train['Age'] = train['DAYS_BIRTH'] // 365
train['DAYS_BIRTH_m'] = np.floor(train['DAYS_BIRTH'] / 30) - ((np.floor(train['DAYS_BIRTH'] / 30) / 12).astype(int) * 12)
train['DAYS_BIRTH_w'] = np.floor(train['DAYS_BIRTH'] / 7) - ((np.floor(train['DAYS_BIRTH'] / 7) / 4).astype(int) * 4)

test['Age'] = test['DAYS_BIRTH'] // 365
test['DAYS_BIRTH_m'] = np.floor(test['DAYS_BIRTH'] / 30) - ((np.floor(test['DAYS_BIRTH'] / 30) / 12).astype(int) * 12)
test['DAYS_BIRTH_w'] = np.floor(test['DAYS_BIRTH'] / 7) - ((np.floor(test['DAYS_BIRTH'] / 7) / 4).astype(int) * 4)
```

##### [EMPLOYED], [DAYS_EMPLOYED_m], [DAYS_EMPLOYED_w]  
EMPLOYED(근속연수), DAYS_EMPLOYED_m(고용된 달) ,DAYS_EMPLOYED_w(고용된 주(고용연도의 n주차)) 피쳐 생성

- DAYS_EMPLOYED_m : month
- DAYS_EMPLOYED_w : week

```python
train['EMPLOYED'] = train['DAYS_EMPLOYED'] // 365
train['DAYS_EMPLOYED_m'] = np.floor(train['DAYS_EMPLOYED'] / 30) - ((np.floor(train['DAYS_EMPLOYED'] / 30) / 12).astype(int) * 12)
train['DAYS_EMPLOYED_w'] = np.floor(train['DAYS_EMPLOYED'] / 7) - ((np.floor(train['DAYS_EMPLOYED'] / 7) / 4).astype(int) * 4)

test['EMPLOYED'] = test['DAYS_EMPLOYED'] // 365
test['DAYS_EMPLOYED_m'] = np.floor(test['DAYS_EMPLOYED'] / 30) - ((np.floor(test['DAYS_EMPLOYED'] / 30) / 12).astype(int) * 12)
test['DAYS_EMPLOYED_w'] = np.floor(test['DAYS_EMPLOYED'] / 7) - ((np.floor(test['DAYS_EMPLOYED'] / 7) / 4).astype(int) * 4)
```

##### [ability]
소득/(살아온 일수+ 근무일수)

```python
train['ability'] = train['income_total'] / (train['DAYS_BIRTH'] + train['DAYS_EMPLOYED'])

test['ability'] = test['income_total'] / (test['DAYS_BIRTH'] + test['DAYS_EMPLOYED'])
```

##### [income_mean]
소득/ 가족 수

```python
train['income_mean'] = train['income_total'] / train['family_size']

test['income_mean'] = test['income_total'] / test['family_size']
```

위와 같이 기존 변수를 기반으로 파생된 변수들을 생성하여 예측에 유의미한 피쳐들을 생성할 수 있으며,

## 5. Scaling, Encoding

다음 기법은 데이터 전처리의 중요한 단계로 모델의 정확성과 일반화 능력을 향상시키는데 기여합니다.  

**Scaling (스케일링):** 데이터의 범위를 조정하여 각 피처의 값들을 일정한 범위로 조절하는 과정입니다. 주로 표준화(Z-score normalization)나 최소-최대 스케일링을 사용하여 모델의 성능을 향상시키고 수렴 속도를 높입니다.

**Encoding (인코딩):** 범주형 데이터를 수치형 데이터로 변환하는 과정입니다. 원-핫 인코딩, 레이블 인코딩 등이 있으며, 머신러닝 알고리즘에서 범주형 데이터를 처리하기 위해 사용됩니다. 인코딩은 범주 간의 순서나 중요도를 부여하지 않으며 모델 성능 향상을 위해 필수적입니다.

```python
numerical_feats = train.dtypes[train.dtypes != "object"].index.tolist()
numerical_feats.remove('credit')
print("Number of Numerical features: ", len(numerical_feats))

categorical_feats = train.dtypes[train.dtypes == "object"].index.tolist()
print("Number of Categorical features: ", len(categorical_feats))
```

위 코드는 수치형 피쳐와 범주형 피쳐의 갯수를 확인하기 위한 코드입니다.

```python
train['income_total'] = np.log1p(train['income_total'])

test['income_total'] = np.log1p(test['income_total'])
```

위 코드는 income_total 열의 변수들을 log변환 시켜주는 코드입니다.

데이터 분석을 하기 위해 log를 취하는 이유는 한마디로 정규성을 높이고 분석(회귀분석 등)에서 정확한 값을 얻기 위함입니다.

데이터 간 편차를 줄여 왜도(skewness)와 첨도(Kurtosis)를 줄일 수 있기 때문에 정규성이 높아지기에 좋은 성능을 내게할 수 있습니다.  

- 해당 코드에서 log대신 log1p를 사용한 이유는 -inf값의 결과를 취하지 않기 위해 1을 더해준 log1p를 사용했습니다.

```python
from sklearn.preprocessing import LabelEncoder
for ord in categorical_feats:
    le = LabelEncoder()
    le.fit(train[ord].astype(str))
    train[ord] = le.transform(train[ord].astype(str))
    
    prev_class = list(le.classes_)
    
    for label in np.unique(test[ord].astype(str)):
        if label not in prev_class:
            prev_class.append(label)
            
    le.classes_ = np.array(prev_class)
    test[ord] = le.transform(test[ord].astype(str))
```

위 코드는 범주형 피쳐들에 대해 LabelEncoding을 하기 위한 코드입니다.
- 라벨인코딩 설명

```python
from sklearn.preprocessing import StandardScaler
numerical_feats.remove('income_total')
scaler = StandardScaler()
train[numerical_feats] = scaler.fit_transform(train[numerical_feats])
test[numerical_feats] = scaler.transform(test[numerical_feats])
```

이후 수치형 변수들에 대해 Standard Scaling을 진행해줍니다.
- Scaling 설명
기존의 log변환한 income_total을 제외한 나머지 수치형 변수들에 대해 Standard Scaling을 적용합니다.  
train에는 fit_transform, test에는 transform

# <span style = 'background-color: #ffdce0'> Stage 6. Tuning & Ensemble

다음은 마지막 단계인 tuning과 ensemble에 대해 진행해보겠습니다. 우선 tuning과 ensemble이 뭔지, 왜 필요한지 짚고 넘어가보도록 하겠습니다.

#### Tuning (하이퍼파라미터 튜닝):   
모델의 성능과 학습 속도에 영향을 주는 하이퍼파라미터를 최적으로 조정하는 과정입니다.   
목적 함수 값을 최소화하거나 최대화하는 방향으로 하이퍼파라미터 조합을 선택해 모델을 개선시킵니다.   
하이퍼파라미터 튜닝은 모델의 최적 설정을 찾아 성능을 극대화하고 과적합을 줄이는 역할을 합니다.

#### Ensemble (앙상블):   
여러 다른 모델을 결합하여 하나의 강력한 모델을 생성하는 기법으로 다양한 모델의 예측을 조합하므로 예측 성능 향상과 과적합 감소를 도모합니다.  
앙상블은 여러 모델의 강점을 결합하여 더 일반화된 예측을 하며, 더 신뢰성 있는 결과를 도출해 더 나은 모델을 구축해준다는 특징을 지닙니다.

# 1. Tuning

tuning을 하기 위해서는 아래와 같은 라이브러리를 불러와야 합니다.

```python
from sklearn.model_selection import StratifiedKFold
from optuna.samplers import TPESampler
from optuna import Trial
import optuna
```

`from sklearn.model_selection import StratifiedKFold`: Scikit-learn 라이브러리의 `StratifiedKFold` 클래스를 임포트합니다.   
이 클래스는 계층화된(k-fold) 교차 검증을 수행할 때 사용되며, 데이터를 일정한 비율로 나누어 모델을 검증하고 평가하는 데 활용됩니다.


`from optuna.samplers import TPESampler`: Optuna 라이브러리의 `TPESampler` 클래스를 임포트합니다.   
이 클래스는 베이지안 최적화를 위한 샘플러 중 하나로, 새로운 하이퍼파라미터 값을 탐색하기 위해 확률적으로 효과적인 접근 방식을 제공합니다.


`from optuna import Trial`: Optuna 라이브러리의 `Trial` 클래스를 임포트합니다.   
`Trial` 클래스는 Optuna의 최적화 과정에서 각각의 시도에 대한 정보와 상태를 담고 있는 객체로, 목적 함수 내에서 이 객체를 활용하여 매개변수 값을 추천하고 결과를 보고할 수 있습니다.

-------------------------------

그럼 CatboostClassifier와 XGB에 대해서 tuning을 진행해보겠습니다.  
tuning을 할 때 함수에 내부 하이퍼 파리미터의 인자값을 넣어주어 정의해주면 됩니다.  
또한 평가지표는 log_loss이기 때문에 log_loss를 사용했습니다

# 1-1. CatBoostClassifier

```python
def objective(trial: Trial) -> float:

    params_catboost = {
        'iterations': trial.suggest_int('iterations', 50, 500),
        'depth': trial.suggest_int('depth', 1, 9),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 1.0),
        'min_child_samples': trial.suggest_int('min_child_samples', 1, 10),
        'border_count': trial.suggest_int('border_count', 1, 255),
        'l2_leaf_reg': trial.suggest_loguniform('l2_leaf_reg', 1e-8, 1.0),
        'eval_metric': 'MultiClass',
        'verbose': False,
    }
    
    X_train, X_valid, y_train, y_valid = train_test_split(train,target,test_size=0.3, random_state=seed)
    
    model = CatBoostClassifier(**params_catboost)
    model.fit(X_train, y_train, eval_set=(X_valid, y_valid), early_stopping_rounds=100, verbose=False)
    
    pred = model.predict_proba(X_valid)
    log_score = log_loss(y_valid, pred)
    
    return log_score
```

이 때, n_trial은 몇 번의 탐색 시도를 진행할지를 나타내는 작업입니다. 해당 실험에서는 300번의 하이퍼파라미터 탐색 시도를 하였습니다.  

```python
sampler = TPESampler(seed=42)
study = optuna.create_study(
    study_name = 'catboost_parameter_opt',
    direction='minimize',
    sampler = sampler,)

study.optimize(objective, n_trials=300) #epoch

print("Best Score:", study.best_value)
print("Best trial:", study.best_trial.params)
```

### 튜닝 후 최종 모델

튜닝을 진행하고 난 후, 최적의 파라미터 값들이 best_params에 저장됩니다.  
최적의 파라미터를 설정하고, test를 예측하는 방식으로 진행해줍니다.    

이 때, 교차검증의 경우 StratifiedKFold를 사용하였습니다.  
참고로 StratifiedKFold는 클래스의 분포를 유지하며 데이터를 나누는 교차 검증 방법을 말합니다.  
많이 실험해본 결과 15개였을 때, 가장 성능이 잘나와 15 fold로 분할해서 진행해주었습니다.

```python
best_params = study.best_trial.params
```

```python
cat_model = CatBoostClassifier(random_state=42, **best_params)

cv = StratifiedKFold(n_splits=15, shuffle=True) 

cr_list = []
sub=np.zeros((test.shape[0], 3)) 

for train_index, test_index in cv.split(train, target) :
        x_train, x_test = train.iloc[list(train_index)], train.iloc[list(test_index)]
        y_train, y_test = target.iloc[list(train_index)], target.iloc[list(test_index)]
        model = cat_model.fit(x_train,y_train)
        pred = model.predict_proba(x_test)
        cross_entropy = log_loss(y_test, pred)
        cr_list.append(cross_entropy)
        sub +=model.predict_proba(test)
        
mysubmission = sub/15

score = sum(cr_list)/15
```

```python
print('logloss:', score)
```

결과로는 0.704 정도가 나왔습니다.

```python
def plot_feature_importance(importance,names,model_type):
    
    feature_importance = np.array(importance)
    feature_names = np.array(names)
    
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)

    plt.figure(figsize=(10,8))

    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])

    plt.title(model_type + ' Feature Importance')
    plt.xlabel('Feature Importance')
    plt.ylabel('Feature Names')

plot_feature_importance(model.get_feature_importance(),test.columns,'CATBOOST')
```

# 1-2. XGBClassifier

다음은 XGBClassifier에 대해 tuning을 진행하겠습니다.

```python
def objective(trial: Trial) -> float:

    params_xgb = {
        'max_depth': trial.suggest_int('max_depth', 1, 9),
        'learning_rate': trial.suggest_loguniform('learning_rate', 0.01, 1.0),
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_loguniform('gamma', 1e-8, 1.0),
        'subsample': trial.suggest_loguniform('subsample', 0.01, 1.0),
        'colsample_bytree': trial.suggest_loguniform('colsample_bytree', 0.01, 1.0),
        'reg_alpha': trial.suggest_loguniform('reg_alpha', 1e-8, 1.0),
        'reg_lambda': trial.suggest_loguniform('reg_lambda', 1e-8, 1.0),
        'eval_metric': 'mlogloss',
        'use_label_encoder': False,
    }
    
    X_train, X_valid, y_train, y_valid = train_test_split(train,target,test_size=0.3, random_state=seed)
    
    model = XGBClassifier(**params_xgb)
    model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], early_stopping_rounds=100, verbose=False)
    
    pred = model.predict_proba(X_valid)
    log_score = log_loss(y_valid, pred)
    
    return log_score
```

마찬가지로 300번의 시도를 하도록 n_trials를 300으로 설정해주었습니다.

```python
sampler = TPESampler(seed=42)
study = optuna.create_study(
    study_name = 'xgbclassifier_parameter_opt',
    direction='minimize',
    sampler = sampler,)

study.optimize(objective, n_trials=300) #epoch

print("Best Score:", study.best_value)
print("Best trial:", study.best_trial.params)
```

### 튜닝 후 최종 모델

```python
best_params = study.best_trial.params
```

```python
xgb_model = XGBClassifier(random_state=42, **best_params)

cv = StratifiedKFold(n_splits=15, shuffle=True) 

cr_list = []
sub=np.zeros((test.shape[0], 3)) 

for train_index, test_index in cv.split(train, target) :
        x_train, x_test = train.iloc[list(train_index)], train.iloc[list(test_index)]
        y_train, y_test = target.iloc[list(train_index)], target.iloc[list(test_index)]
        model = xgb_model.fit(x_train,y_train)
        pred = model.predict_proba(x_test)
        cross_entropy = log_loss(y_test, pred)
        cr_list.append(cross_entropy)
        sub +=model.predict_proba(test)
        
mysubmission = sub/15

score = sum(cr_list)/15
```

# 1-3. 추가 RandomForestClassifier

모델 간의 비교를 위해 추가적으로 Tree 기반 모델에 대해 tuning을 한 번 더 진행하겠습니다.

```python
from sklearn.ensemble import RandomForestClassifier

def objective(trial: Trial) -> float:
    params_rf = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 500),
        'max_depth': trial.suggest_int('max_depth', 1, 32),
        'min_samples_split': trial.suggest_float('min_samples_split', 0.1, 1.0),
        'min_samples_leaf': trial.suggest_float('min_samples_leaf', 0.1, 0.5),
        'max_features': trial.suggest_categorical('max_features', ['auto', 'sqrt', 'log2']),
        'bootstrap': trial.suggest_categorical('bootstrap', [True, False]),
        'random_state': 42, 
    }

    X_train, X_valid, y_train, y_valid = train_test_split(train,target,test_size=0.3, random_state=seed)
    
    model = RandomForestClassifier(**params_rf)
    model.fit(X_train, y_train)
    
    pred = model.predict_proba(X_valid)
    log_score = log_loss(y_valid, pred)
    
    return log_score
```

```python
sampler = TPESampler(seed=42)
study = optuna.create_study(
    study_name = 'rfclassifier_parameter_opt',
    direction='minimize',
    sampler = sampler)

study.optimize(objective, n_trials=300) #epoch

print("Best Score:", study.best_value)
print("Best trial:", study.best_trial.params)
```

### 튜닝 후 최종 모델

```python
best_params = study.best_trial.params
```

```python
rf_model = RandomForestClassifier(random_state=42, **best_params)

cv = StratifiedKFold(n_splits=15, shuffle=True) 

cr_list = []
sub=np.zeros((test.shape[0], 3)) 

for train_index, test_index in cv.split(train, target) :
        x_train, x_test = train.iloc[list(train_index)], train.iloc[list(test_index)]
        y_train, y_test = target.iloc[list(train_index)], target.iloc[list(test_index)]
        model = rf_model.fit(x_train,y_train)
        pred = model.predict_proba(x_test)
        cross_entropy = log_loss(y_test, pred)
        cr_list.append(cross_entropy)
        sub +=model.predict_proba(test)
        
mysubmission = sub/15

score = sum(cr_list)/15
```

Randomforest보다는 Catboost와 XGBclassifier의 성능이 더 좋은 것을 확인할 수 있습니다.

위의 작업을 최종 제출물로 제출할 수도 있지만, **ensemble**이라는 기법을 추가적으로 사용하여 성능을 더 끌어올릴 수 있습니다.  


위해서 언급했듯이 앙상블은   
여러 다른 모델을 결합하여 하나의 강력한 모델을 생성하는 기법으로 다양한 모델의 예측을 조합하므로 예측 성능 향상과 과적합 감소를 도모합니다. 
또한, 앙상블은 여러 모델의 강점을 결합하여 더 일반화된 예측을 하며, 더 신뢰성 있는 결과를 도출해 더 나은 모델을 구축해준다는 특징을 지닙니다.

## 2-1. 가중평균 앙상블

앙상블의 종류는 많지만, 그 중 간단하게 진행할 수 있는 **가중 평균 앙상블**을 진행해보려고 합니다. 

가중평균 앙상블은 어떻게 진행되는지 알아보고 가겠습니다.  

 
**가중평균 앙상블**이란, 다수의 모델 예측 결과를 합치는데 각 모델의 예측 결과에 가중치를 부여하여 최종 예측을 생성하는 앙상블 방법입니다.   
이 가중치는 모델의 성능이나 신뢰도에 따라 조정되며, 모델 간 상대적 중요도를 반영하여 더 효과적인 예측을 할 수 있도록 도와줍니다.

우선 빈 리스트를 생성해서 각 csv로 나온 파일들을 더해줍니다.

```python
clfs = []
cat = pd.read_csv('submission_CatBoost0.7040.csv') ; clfs.append(cat)
xgb= pd.read_csv('submission_XGB0.6868.csv') ; clfs.append(xgb)
rf= pd.read_csv('submission_rf0.8525.csv') ; clfs.append(rf)
```

id 설정을 위해 아래와 같이 iloc을 사용해주었습니다. 또한 log_loss 리스트 안에 해당 성능을 적어줍니다. 

```python
data_id = cat.iloc[:,0]
log_loss = [0.7040, 0.6868,0.8525 ]
```

상관관계를 찍어서 각 모델끼리의 상관성을 파악할 수 있습니다. 이 때 주의해야 할 점은 상관관계가 서로 높은 모델들은 **지양**해야 한다는 점입니다.   
이유로는 다음과 같습니다. 

- **다중공선성 문제**: 상관관계가 높은 모델들을 함께 사용하면, 모델들이 서로 유사한 예측을 하게 되므로 불필요한 중복 정보가 증가할 수 있습니다. 이는 모델 간 다중공선성 문제를 야기하여 가중평균 앙상블의 안정성과 신뢰도를 감소시킬 수 있습니다.  


- **과적합 가능성**: 상관관계가 높은 모델들을 함께 사용하면, 훈련 데이터에 과적합되어 예측 성능이 개선되는 대신 새로운 데이터에 대한 일반화 능력이 저하될 수 있습니다.  


- **다양성 부족**: 앙상블은 다양한 모델들의 예측 결과를 결합하여 성능을 향상시키는 것이 목표입니다. 상관관계가 높은 모델들을 사용하면 다양성이 부족해질 가능성이 높아져서 앙상블의 효과를 충분히 누리지 못할 수 있습니다.  


```python
genders = []
for i in range(3):
    genders.append(clfs[i].iloc[:,-1])

ensemble_results = pd.concat(genders, axis=1)
col = ['cat','xgb','rf']
ensemble_results.columns = col
plt.figure(figsize = (12,9))
g = sns.heatmap(ensemble_results.corr(),annot=True, cmap='Blues')
g.set_title("Correlation between models")
plt.show()
```

catboost와 xgboost의 상관관계는 0.96을 보여주고 있으며, randomforest 모델은 예측간의 상관관계가 상당히 낮은 모습을 보여주고 있습니다.

추가적으로 아래 코드를 활용해 시각화할 수도 있습니다!  
아래의 코드는 여러 모델의 예측 결과를 분석하여 상관관계와 예측 성능(log_loss) 간의 관계를 시각화하여, 서로 상관되지 않은 모델들의 조합이 예측 결과의 다양성을 향상시킬 수 있는지를 분석하는 데 사용될 수 있습니다.  

x축의 경우, 상관관계의 평균값을 나타내며, y축은 log_loss의 값을 나타냅니다. 

```python
corr = (ensemble_results.corr().sum()-1)/(ensemble_results.corr().shape[0]-1)
names = corr.index.str[:]
aucs = [-x for x in log_loss]
df = pd.DataFrame({'model': names, 'auc': aucs, 'cor': corr})        

plt.figure(figsize=(12,10))
g = sns.scatterplot(x="cor", y="auc", data=df, s=40, color='red')
for line in range(0, df.shape[0]):
     g.text(df.cor[line]+0.001, df.auc[line]-0.001, 
            df.model[line], horizontalalignment='left', 
            size='medium', color='black', weight='semibold')
        
plt.xlim((df.cor.min()-0.01,df.cor.max()+0.01))
plt.ylim((df.auc.min()-0.01,df.auc.max()+0.01))
plt.xlabel('Mean Agreement')
plt.ylabel('log_loss')
plt.grid()
plt.show()
```

submission 앙상블 시 상관관계가 낮은 모델끼리의 앙상블이 효과가 좋지만 randomforest 모델의 성능이 나머지 두 모델에 비해 너무 낮기 때문에 catboost, xgboost 두 모델에 대해서만 앙상블 해보도록 하겠습니다.

두 모델 중 xgboost가 더 좋기 때문에 xgboost model에 가중치를 더 주어 앙상블 해보겠습니다.  
참고로 가중평균 앙상블의 경우, 가중치를 각각 다르게 설정하면 성능도 그에 따라 바뀌기 때문에, 많은 실험을 해보시는 것을 추천드립니다.
